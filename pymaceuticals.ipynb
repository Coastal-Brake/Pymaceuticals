```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_complete = pd.merge(study_results, mouse_metadata, how="right", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
study_complete.head(50)
```

```python
# Checking the number of mice.
study_complete["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
#index_study_complete = study_complete.set_index('Mouse ID')
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
index_set_df = study_complete.set_index(["Mouse ID"])
index_set_df.loc['g989']
```

```python

# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_df = index_set_df.drop(["g989"])
drop_that_mouse=index_set_df.drop_duplicates(subset=None, keep='first', inplace=False)
#drop_that
squeaky_clean_df=drop_that_mouse.reset_index(['Mouse ID'])
```

```python
# Checking the number of mice in the clean DataFrame.
squeaky_clean_df["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
tumor = squeaky_clean_df.groupby(['Drug Regimen'])
 #Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
 #mean, median, variance, standard deviation, and SEM of the tumor volume.
tumor_mean=tumor["Tumor Volume (mm3)"].mean()
tumor_median=tumor["Tumor Volume (mm3)"].median()
tumor_variance=tumor["Tumor Volume (mm3)"].var()
tumor_std=tumor["Tumor Volume (mm3)"].std()
tumor_sem=tumor["Tumor Volume (mm3)"].sem()
# Assemble the resulting series into a single summary DataFrame.
summary_df=pd.DataFrame(
        {"Mean Tumor Volume":tumor_mean,
         "Median Tumor Volume":tumor_median,
         "Tumor Volume Variance":tumor_variance,
         "Tumor Volume Std. Dev.":tumor_std,
         "Tumor Volume Std. Err.": tumor_sem})
summary_df.head(10)
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
# Calculate mean, median, and standard deviation for the 'Value' column

agg_sum = index_set_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

#print(agg_sum)

# Create a summary DataFrame using the aggregated values
summary_df = agg_sum.rename(columns={
    'mean': 'Mean',
    'median': 'Median',
    'var': 'Variance',
    'std':  "Std",
    'sem': 'SEM'
}).reset_index()
summary_df.head(20)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
####setting to max is important, because although 45 is usually the last timepoint, filtering by that doesn't
####catch the mice that don't make it that far
#narrow down to final timepoint
final_tp = squeaky_clean_df.groupby(['Mouse ID'])['Timepoint'].max() 
final_tp_df=pd.DataFrame(final_tp)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_vol_df = pd.merge(squeaky_clean_df, final_tp_df, how='right', on=['Mouse ID','Timepoint'])
final_vol_df.head(40)
```

```python
# Create a list of the target regimens
target_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Filter the DataFrame to include only the specified regimens
filtered_df = final_vol_df[final_vol_df["Drug Regimen"].isin(target_regimens)]
filtered_df.head()
```

```python
#Put treatments into a list for for loop (and later for plot labels)
#reduced the code significantly by using a function 

def potential_outliers(pd_series, column_name):
    series_avg=pd_series.mean
    q1 = pd_series.quantile(0.25)
    q3 = pd_series.quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    outliers = pd_series[(pd_series < lower_bound) | (pd_series > upper_bound)]
    print(f"{column_name}'s potential outliers: {outliers}")



treatment=["Capomulin", "Ramicane", "Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
cap_vol=[]
ram_vol=[]
inf_vol=[]
cef_vol=[]

for index, row in final_vol_df.iterrows():
    if row["Drug Regimen"] == treatment[0]:
        cap_vol.append(row["Tumor Volume (mm3)"])
    elif row["Drug Regimen"] == treatment[1]:
        ram_vol.append(row["Tumor Volume (mm3)"])
    elif row["Drug Regimen"] == treatment[2]:
        inf_vol.append(row["Tumor Volume (mm3)"])
    elif row["Drug Regimen"] == treatment[3]:
        cef_vol.append(row["Tumor Volume (mm3)"])


        
cap_vol_series = pd.Series(cap_vol, name=treatment[0])
ram_vol_series = pd.Series(ram_vol, name=treatment[1])
inf_vol_series = pd.Series(inf_vol, name=treatment[2])
cef_vol_series = pd.Series(cef_vol, name=treatment[3])

potential_outliers(cap_vol_series,"Capomulin") 
potential_outliers(ram_vol_series,"Ramicane")
potential_outliers(inf_vol_series,"Infubinol")
potential_outliers(cef_vol_series,"Ceftamin")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
#ax = reg_vol_df.boxplot(column=["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
#ax.grid(False)  # Set the grid to False to remove grid lines
#plt.ylabel("Final Tumor Volume (mm3)")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8,6))

# Create a boxplot for the desired column with custom colors
columns_to_plot = [cap_vol_series, ram_vol_series, inf_vol_series, cef_vol_series]
flierprops = dict(markerfacecolor='red',markersize=12)
labels = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

boxplot = ax.boxplot(columns_to_plot, patch_artist=True,flierprops=flierprops, labels=labels)

for box in boxplot['boxes']:
    box.set(facecolor='white')  # You can change 'white' to the desired color


# Customize the appearance
ax.grid(False)
ax.set_ylabel("Final Tumor Volume (mm3)")
# Show the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_clean = squeaky_clean_df.loc[squeaky_clean_df["Drug Regimen"] == "Capomulin"]
cap_mt_tavg= capomulin_clean.groupby(['Mouse ID','Weight (g)',"Drug Regimen"])["Tumor Volume (mm3)"].mean().reset_index()
#Rename "Tumor Volume (mm3)" to 'Average Tumor Volume (mm3)' to be correct
cap_mt_tavg = cap_mt_tavg.rename(columns={'Tumor Volume (mm3)': 'Average Tumor Volume (mm3)'})
cap_mt_tavg_df=pd.DataFrame(cap_mt_tavg)

x = cap_mt_tavg_df['Weight (g)']
y = cap_mt_tavg_df['Average Tumor Volume (mm3)']
plt.scatter(x,y)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
x = cap_mt_tavg_df['Weight (g)']
y = cap_mt_tavg_df['Average Tumor Volume (mm3)']

correlation_coefficient, _ = st.pearsonr(x,y)
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x,y)
regress_values = x * slope + intercept
print(f"The correlation between mouse weight and the average tumor volume is: {correlation_coefficient:.2f}")

plt.plot(x, slope * x + intercept, color='red', label=f'Regression Line: y = {slope:.2f}x + {intercept:.2f}')
plt.scatter(x, y)
plt.annotate(f'Correlation: {correlation_coefficient:.2f}', (5.8, 0.8), fontsize=15, color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```
